```python
def calculate_the_volume_of_sphere(radius):
    volume = 4./3*np.pi*(radius**3)
    surface = np.pi*(radius**2)
    return {'a':volume,'b':surface}
    #return volume, surface
```

```python
r = 5
output = calculate_the_volume_of_sphere(r)

print(output['a'])
print(output['b'])

#print(volume,surface)
```

```python

def my_plot_style():
   
    params = {'legend.fontsize': 20,
             'axes.labelsize': 20,
             'axes.titlesize':20,
             'xtick.labelsize':20,
             'ytick.labelsize':20,
             'xtick.major.size':5,
              'xtick.minor.size':2.5,
             'ytick.major.size':5,
              'ytick.minor.size':2.5,
             'figure.facecolor':'w',
             #'lines.linewidth' : 1.5,
              'xtick.major.width':1.5,
              'ytick.major.width':1.5,
              'xtick.minor.width':1.5,
              'ytick.minor.width':1.5,
              'axes.linewidth':1.5,
              'xtick.direction':'in',
              'ytick.direction':'in',
             'ytick.labelleft':True,
              'text.usetex' : False,
             'font.family': 'sans-serif',
              #'font.family': 'fantasy'
             }
  
    plt.rcParams.update(params)
```

```python
def removing_ytick_label():
    ax = plt.gca()
    ax.axes.yaxis.set_ticklabels([])
```

```python
my_plot_style()
plt.figure(figsize=(10,10))
noise = np.random.normal(0,0.5,len(x))
plt.plot(x,y,lw=5,ls=':')
plt.errorbar(x,y+noise,yerr=0.5,xerr=0.5,ls='',color='C0',marker='s',markersize=10,capsize=8,lw=3,mew=3,label='Lan et al. (2020)',zorder=1)
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
```

# Mini exercise - plotting SDSS spectrum

### 1. Please explore the following fits file and produce a similar plot as shown below. 
### 2. What is the redshift of the galaxy? Hint: the strongest emission line is Ha [rest frame 6564.614 ]



```python
import astropy.io.fits as pf
my_plot_style()
spectrum = pf.open('spec-0285-51930-0184.fits')

x = 10**spectrum[1].data['loglam']
y = spectrum[1].data['flux']
y_model = spectrum[1].data['model']
plt.figure(figsize=(15,5))
plt.plot(x,y,lw=0.8)
plt.plot(x,y_model,lw=0.8,color='r')

plt.xlabel('Wavelength [$\\rm \\AA$]')
plt.ylabel('Flux [$\\rm 10^{-17} erg/s/cm^{2}/\\AA $]')


max_y = np.max(y)
search_max = np.where(y==max_y)[0]

print(search_max)
max_x = x[search_max]
print(max_x[0])
plt.axvline(max_x,ls='--',color='red',zorder=10,lw=0.5)
print('Redshift:',max_x[0]/6564.614-1)
plt.ylim(0,30)
estimated_z = max_x[0]/6564.614-1
print('Catalog redshift:',spectrum[2].data['Z'][0])
model_z = spectrum[2].data['Z'][0]
model_zerr = spectrum[2].data['Z_ERR'][0]
print('dz',model_z-estimated_z)
#print('z_err',model_zerr)
```
